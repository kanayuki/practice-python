```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.animation import FuncAnimation
```

```python
x = [1, 2, 3, 4, 5]
y = [1, 2, 4, 3, 5]
fig, axes = plt.subplots(1, 3, figsize=(16, 6))
ax1, ax2, ax3 = axes
ax1.plot(x, y)
ax2.scatter(x, y)
ax3.barh(x, y)
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [1, 2, 4, 3, 5]
plt.figure(figsize=(16, 6))
plt.subplot(1, 3, 1)
plt.plot(x, y)
plt.subplot(1, 3, 2)
plt.scatter(x, y)
plt.subplot(1, 3, 3)
plt.barh(x, y)
plt.show()
```
